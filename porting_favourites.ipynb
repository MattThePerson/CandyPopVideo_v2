```python
import json
from app.schemas import VideoInteractions
from app import db
from datetime import datetime
from datetime import datetime, timedelta
import os
from handymatt_media import video_analyser
```

```python
with open('data/metadata.json', 'r') as f:
    data = json.load(f)
with open('data/videos.json', 'r') as f:
    videos = json.load(f)
```

```python
favourites = data['favourites']
views = data['view_history']
```

```python
favourites
```

```python
time_format = "%Y-%m-%dT%H:%M:%S"
now = datetime.strptime('2025-06-30T18:00:00', time_format)
items = []
for video_hash in favourites:
    now = now + timedelta(minutes=1)
    formatted = now.strftime(time_format)
    video = videos.get(video_hash, {})
    path = video.get('path')
    exists = os.path.exists(path)
    exists_str = "EXISTS" if exists else "no    "
    print('[{}]  |  {}  |  {}  |  "{}"'.format(video_hash, formatted, exists_str, path))
    items.append({
        'hash': video_hash,
        'time_added': formatted,
        'path': path,
        'exists': exists,
    })
```
